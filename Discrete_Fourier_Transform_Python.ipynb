```python
from matplotlib import pyplot as plt
import numpy as np
import math
import cmath
import array as arr

# Mendefinisikan variabel-variabel 
N=500 # Jumlah sample
x = arr.array('f',[0])
y = arr.array('f',[0])
k = arr.array('f',[0])
f_hatr = arr.array('f',[0]) 
f_hati = arr.array('f',[0])
f_hat = arr.array('f',[0])

x=[0]*(N+1)
y=[0]*(N+1)
k=[0]*(N+1)
f_hatr=[0]*(N+1)
f_hati=[0]*(N+1)
f_hat=[0]*(N+1)

x1=-10 # Batas x1
x2=10 # Batas x2
dx=(x2-x1)/N # Interval

# Menentukan fungsi sinusoidal yang akan ditransform
for i in range (0,N,1):
  x[i]=x1+i*dx
  y[i]=np.sin(x[i])+2*np.sin(3*x[i])

plt.scatter(x, y)
plt.show()

#Discrete Fourier Transform
for i in range (0,N,1):
 k[i]=-(N//2)+(i)
 A=0
 B=0
 for j in range (0,N,1):
  A = A + y[j]*(np.cos(k[i]*x[j])*dx)  
  B = B - y[j]*(np.sin(k[i]*x[j])*dx)
 f_hatr[i]=A
 f_hati[i]=B
 f_hat[i]=A*A+B*B

#plot norm
plt.scatter(k,f_hat)
plt.xlim (x1,x2)
plt.xticks(np.arange(x1, x2+1, 1))
plt.ylim (0,15)
plt.yticks(np.arange(0, 100, 15))
plt.stem(k,f_hat)
plt.show() 

#plot bagian real dan imaginer secara terpisah
plt.scatter(k,f_hatr,color='blue', marker='*', label='Real')
plt.scatter(k,f_hati,color= 'red', marker='.', label='Imaginer')
plt.xlim(-15, 15)
plt.legend(loc='best')
plt.show()
```
